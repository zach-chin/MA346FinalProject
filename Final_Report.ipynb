 The goal of this project is to create a model that is capable of predicting the price of avocados of a specified region in the future. This was done by using the Python function `curve_fit` to fit a model to historical data, then using this model to predict prices in the future of the data. During the project I was able to fit the historical data to a sin curve with 5 variables influencing the pattern. I was also able to fit a time series model for prediction, which will be used when it is more accurate. Using these models I was able to predict things such as the price of an organic avocado in Las Vegas being $1.59 in Winter of 2019. Different prices can be predicted here LINK TO DASHBOARD. 

 The data used in this project came from Kaggle where it had been put into a csv file from data downloaded from the Hass Avacoda Board (https://www.kaggle.com/datasets/neuromusic/avocado-prices?resource=download). This fit what I needed becuase it had columns that were easy to filter to find the specific combinations of region and type of avocado. It also has a few more columns of data that could be used for further filtering if needed. To make it suitable for applying to a model it needed to be sorted by ascending date. Then we just needed to filter for the region of interest. The only issue with this data was that it ended in the beginning of 2018, meaning that it wouldn't be very useful for a prediction in 2022 becuase that would be very far extrapolation and less accurate. Therefore I only used it to predict 2019 to demonstrate the ability to predict.

The model that was used for predicting the price of the avocados is ε * SIN(α + β * t) + (γ * t) + ρ, where t is time. ε is a variables that impacts the amplitude of the sin curve, α is for shifting the sin curve left or right, β is for frequency, γ for slope, and ρ is the intercept. For each different region chosen these variables are chosen to select a curve best for the rise and fall of the price of avocados throughout the seasons. The time series model that I used is predicted based off of the fit.model function in the statsmodels package. This took the data of the avocados and used it to create an ARIMA seasonal model to predict values and graph a curve. This tended to be less accurate with the data and I chose to only use it in the dashboard when it had less error than the previous model. In this case the sin curve model was a better fit than the time series model, so if this data is inputted into the dashboard it will choose to present the sin curve. Through the use of these models all of the data in the dataset can be used for future prediction of the price of the avocados in their respective locations.
